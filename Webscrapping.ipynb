```python
def get_and_parse(url):
  response = requests.get(url)

# Parse the HTML content with Beautiful Soup
  soup = BeautifulSoup(response.content, 'html.parser')

# Remove unwanted elements (script, style, comments, etc.)
  for element in soup(['script', 'style', 'comment']):
      element.extract()

# Extract the visible text
  text = soup.get_text(separator=' ')
  return text
```

```python
def parse_elements(url,element):
    session = HTMLSession()
    r = session.get(url)
    elements = r.html.find(element)
    return elements
```

```python
def Words_detection(sentence):
    Important_words = ('About', 'Terms', 'Policy')
    Unimportant_words = ('Advertised', 'Staff', 'Contact')

    Important = any(sentence.count(i) > 0 for i in Important_words)
    Unimportant = any(sentence.count(i) > 0 for i in Unimportant_words)

    if Important == Unimportant == False:
        return "neutral"
    elif Important != Unimportant:
        return "Important" if Important else "Unimportant"
    else:
        return "mixed"
```
